```python
# 6.write a code to perform basic string compression using the counts of repetead characters.

def compress_string(string):
    """Compresses a string using the counts of repeated characters."""

    if not string:
        return string  

    compressed = ""
    count = 1

    for i in range(1, len(string)):
        if string[i] == string[i - 1]:
            count += 1
        else:
            compressed += string[i - 1] + str(count)
            count = 1

    
    compressed += string[-1] + str(count)

    return compressed if len(compressed) < len(string) else string


string = "aabcccccaaa"
print(compress_string(string))  # Output: "a2b1c5a3"
```

```python
# 7.write a code to determine if a string has all unique characters.

def has_all_unique_characters(s):

    seen_characters = set()
    

    for char in s:
      
        if char in seen_characters:
            return False
     
        seen_characters.add(char)
    
    
    return True


print(has_all_unique_characters("abcdef"))  # True
print(has_all_unique_characters("hello")) #false
```

```python
# 25. Create a code that takes a tuple and two integers as input.The functions should return a new tuple containing elements from the original tuple within the specified range of indices.


def slice_tuple(original_tuple, start_index, end_index):
    """
    Return a new tuple containing elements from the original tuple within the specified range of indices.

    Parameters:
    original_tuple (tuple): The tuple from which to extract elements.
    start_index (int): The starting index of the range (inclusive).
    end_index (int): The ending index of the range (exclusive).

    Returns:
    tuple: A new tuple containing elements within the specified range.
    """
    
    if start_index < 0:
        start_index = 0
    if end_index > len(original_tuple):
        end_index = len(original_tuple)
    if start_index > end_index:
        raise ValueError("start_index cannot be greater than end_index")


    sliced_tuple = original_tuple[start_index:end_index]
    return sliced_tuple


original = (1, 2, 3, 4, 5, 6, 7, 8, 9)
start = 2
end = 5
result = slice_tuple(original, start, end)
print(result)  # Output: (3, 4, 5)
```

```python
# 33.Write a code to access a value in a nested dictionary.The function should take the dictionary and a list of keys as input, and return the corresponding value.if any of the keys do not exist in the dictionary,the funtion should return None.



def get_nested_value(d, keys):
    """
    Accesses a value in a nested dictionary using a list of keys.

    Args:
    d (dict): The nested dictionary to search.
    keys (list): A list of keys to access the value.

    Returns:
    The value corresponding to the list of keys, or None if any key is not found.
    """
    current = d
    for key in keys:
        if isinstance(current, dict) and key in current:
            current = current[key]
        else:
            return None
    return current

# Example usage
nested_dict = {
    'a': {
        'b': {
            'c': 10
        }
    }
}


keys = ['a', 'b', 'c']
value = get_nested_value(nested_dict, keys)

print("Value:", value)  # Output: Value: 10


keys_invalid = ['a', 'b', 'd']
value_invalid = get_nested_value(nested_dict, keys_invalid)

print("Value for invalid keys:", value_invalid)  # Output: Value for invalid keys: None
```

```python
# 34.Write a code that takes a dictionary as input and returns a sorted version of it based on the values. You can choose whether to sort in ascending or decending order. 


def sort_dict_by_value(d, ascending=True):
    """
    Sorts a dictionary by its values.

    Args:
    d (dict): The dictionary to sort.
    ascending (bool): If True, sorts in ascending order. If False, sorts in descending order.

    Returns:
    dict: A new dictionary sorted by values.
    """

    sorted_items = sorted(d.items(), key=lambda item: item[1], reverse=not ascending)
    
  
    sorted_dict = dict(sorted_items)
    
    return sorted_dict

# Example usage
input_dict = {'a': 3, 'b': 1, 'c': 2}
sorted_dict_ascending = sort_dict_by_value(input_dict, ascending=True)
sorted_dict_descending = sort_dict_by_value(input_dict, ascending=False)

print("Sorted in ascending order:", sorted_dict_ascending)
print("Sorted in descending order:", sorted_dict_descending) #Sorted in ascending order: {'b': 1, 'c': 2, 'a': 3}
#Sorted in descending order: {'a': 3, 'c': 2, 'b': 1}
```

```python
# 35. Write a code that a dictionary, swapping keys and values. Ensure that the inverted dictionary correctly handles cases where multiple keys have the same value by storing the keys as a list in the inverted dictionary.


def invert_dict(d):
    """
    Swaps keys and values in a dictionary. 
    If multiple keys have the same value, the inverted dictionary stores the keys as a list.

    Args:
    d (dict): The dictionary to invert.

    Returns:
    dict: A new dictionary with keys and values swapped.
    """
    inverted_dict = {}
    
    for key, value in d.items():
        if value in inverted_dict:
            
            if isinstance(inverted_dict[value], list):
                inverted_dict[value].append(key)
            else:
                inverted_dict[value] = [inverted_dict[value], key]
        else:
            
            inverted_dict[value] = key
    
    return inverted_dict


original_dict = {
    'a': 1,
    'b': 2,
    'c': 1,
    'd': 3
}

inverted = invert_dict(original_dict)
print("Inverted dictionary:", inverted) #Inverted dictionary: {1: ['a', 'c'], 2: 'b', 3: 'd'}
```
